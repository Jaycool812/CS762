```python
import pandas as pd
import numpy as np
```

```python
#view the data
data = pd.read_csv("../data.csv",header=None)
print(data.shape)
data
```

```python
from sklearn.model_selection import train_test_split
col_nums = data.shape[1]
X = data.iloc[:,0:col_nums-1]
y = data.iloc[:,col_nums-1:col_nums]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=1)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## First try filling NA with a constant

```python
X_train1 = X_train.fillna(-100000)
```

```python
from sklearn import tree
from sklearn.model_selection import cross_val_score
clf = tree.DecisionTreeClassifier()
accuracy = cross_val_score(clf,X_train1,y_train,scoring="accuracy",cv=10)
print(f"Accuracy for each cv is:{accuracy}")
print("---------------------------")
print(f"Mean accuracy is {np.mean(accuracy)}")
```

```python
#Accuracy on test set
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train1,y_train)
X_test1 = X_test.fillna(-100000)
y_pred = clf.predict(X_test1)
clf.score(X_test1,y_test)
```

## Then we try filling NA with mean

```python
X_train2 = X_train.apply(lambda x:x.fillna(x.mean()),axis=0)
y_train2 = y_train.fillna(np.mean(y_train))
```

```python
X_train2
```

```python
clf = tree.DecisionTreeClassifier()
accuracy = cross_val_score(clf,X_train2,y_train2,scoring="accuracy",cv=10)
print(f"Accuracy for each cv is:{accuracy}")
print("---------------------------")
print(f"Mean accuracy is {np.mean(accuracy)}")
```

```python
#Accuracy on test set
X_test2 = X_test.fillna(X_train.mean())
clf = tree.DecisionTreeClassifier(random_state = 1)
clf = clf.fit(X_train2,y_train)
y_pred = clf.predict(X_test2)
clf.score(X_test2,y_test)
```
